```python
!pip install transformers

import json
from transformers import pipeline
from IPython.display import display, HTML, Javascript
import ipywidgets as widgets

# Sample corpus
corpus = {
    "What is your best red wine?": "Our best red wine is the 2015 Cabernet Sauvignon.",
    "Tell me more about it.": "The 2015 Cabernet Sauvignon is rich and full-bodied, with flavors of dark fruit and a hint of oak."
}

# Create a retrieval-based chatbot function
def chatbot_response(question):
    if question in corpus:
        return corpus[question]
    else:
        return "Please contact the business directly for this information."

# Create input and output widgets
input_box = widgets.Textarea(
    value='',
    placeholder='Type your question here...',
    description='You:',
    disabled=False,
    layout=widgets.Layout(width='80%', height='80px')
)

output_box = widgets.HTML(
    value='',
    placeholder='',
    description='Bot:',
    layout=widgets.Layout(width='80%', height='100px')
)

send_button = widgets.Button(
    description='Send',
    disabled=False,
    button_style='',
    tooltip='Send',
    icon='check'
)

# Function to handle button click
def on_button_click(b):
    question = input_box.value
    answer = chatbot_response(question)
    output_box.value += f"<b>You:</b> {question}<br><b>Bot:</b> {answer}<br><br>"
    input_box.value = ''

send_button.on_click(on_button_click)

# Display the widgets
display(input_box, send_button, output_box)
```
